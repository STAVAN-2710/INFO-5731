# Question 4 (10 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
def break_into_parts(expr):
    # Split the expression into smaller parts
    parts = []
    current_part = ''

    for char in expr:
        if char.isspace():
            if current_part:
                parts.append(current_part)
                current_part = ''
        elif char in '+-*/()^%√':  # Allowing basic math symbols and functions
            if current_part:
                parts.append(current_part)
                current_part = ''
            parts.append(char)
        else:
            current_part += char

    if current_part:
        parts.append(current_part)

    return parts

def do_math(num1, num2, op):
    # Perform the math operation based on the operator
    num1, num2 = float(num1), float(num2)

    if op == '+':
        return num1 + num2
    elif op == '-':
        return num1 - num2
    elif op == '*':
        return num1 * num2
    elif op == '/':
        if num2 == 0:
            raise ZeroDivisionError("Can't divide by zero!")
        return num1 / num2
    elif op == '^':
        return num1 ** num2
    elif op == '%':
        if num2 == 0:
            raise ZeroDivisionError("Can't divide by zero!")
        return num1 % num2
    elif op == '√':
        # For square root, only use the first number
        if num1 < 0:
            raise ValueError("Can't find square root of a negative number!")
        return num1 ** 0.5

def check_priority(op):
    # Determine which operation has higher priority
    if op in ['+', '-']:
        return 1
    if op in ['*', '/', '%']:  # Modulus has the same priority as multiplication/division
        return 2
    if op in ['^', '√']:  # Square root has high priority too
        return 3
    return 0

def solve_expression(expr):
    # Set up stacks for numbers and operators
    number_stack = []
    operator_stack = []

    parts = break_into_parts(expr)

    for part in parts:
        if part.replace('.', '').isdigit():  # Check if it's a number
            number_stack.append(float(part))
        elif part in '+-*/^%√':  # Check for operators
            while (operator_stack and
                   check_priority(operator_stack[-1]) >= check_priority(part)):
                current_op = operator_stack.pop()
                if current_op == '√':
                    n = number_stack.pop()
                    number_stack.append(do_math(n, 0, current_op))
                else:
                    second_num = number_stack.pop()
                    first_num = number_stack.pop()
                    number_stack.append(do_math(first_num, second_num, current_op))
            operator_stack.append(part)

    # Finish up with any remaining operations
    while operator_stack:
        current_op = operator_stack.pop()
        if current_op == '√':
            n = number_stack.pop()
            number_stack.append(do_math(n, 0, current_op))
        else:
            second_num = number_stack.pop()
            first_num = number_stack.pop()
            number_stack.append(do_math(first_num, second_num, current_op))

    answer = number_stack[0]

    # Clean up decimal places for nicer output
    return round(answer, 4) if isinstance(answer, float) else answer

# Main calculator program starts here
print("Welcome to the Enhanced Calculator!")
print("You can do the following operations:")
print("  +  Addition")
print("  -  Subtraction")
print("  *  Multiplication")
print("  /  Division")
print("  ^  Exponentiation (like: 2^3)")
print("  %  Modulus (remainder)")
print("  √  Square root (like: √16)")
print("Type 'exit' when you're done")

while True:
   try:
       # Ask for the math problem from the user
       problem = input("\nWhat calculation do you want to perform? ")

       if problem.lower() == 'exit':
           print("Thanks for using the calculator! Goodbye!")
           break

       if not problem.strip():
           continue

       # Solve the expression and show the answer
       result = solve_expression(problem)
       print(f"The answer is: {result}")

   except ZeroDivisionError:
       print("Oops! You can't divide by zero!")
   except ValueError as e:
       print(f"Oops! {str(e)}")
   except Exception as e:
       print("Something went wrong. Please check your input.")
```

```python
import csv
import os
import re
from typing import Dict, List
import threading
from datetime import datetime, timedelta
import random  # For simulating seasonal demand

# Global lock for file operations
file_lock = threading.Lock()

# Simulated data for multi-channel inventory
online_inventory = {}
store_inventory = {}

def create_default_csv():
    default_products = [
        {"ProductID": "P001", "Name": "Laptop", "Category": "Electronics", "Price": "999.99", "Quantity": "10", "RestockLevel": "5", "ExpirationDate": "", "LeadTime": "14"},
        {"ProductID": "P002", "Name": "Desk Chair", "Category": "Furniture", "Price": "199.99", "Quantity": "20", "RestockLevel": "10", "ExpirationDate": "", "LeadTime": "30"},
        # Add more default products here...
    ]
    with open("inventory.csv", "w", newline='') as f:
        writer = csv.DictWriter(f, fieldnames=["ProductID", "Name", "Category", "Price", "Quantity", "RestockLevel", "ExpirationDate", "LeadTime"])
        writer.writeheader()
        writer.writerows(default_products)
    print("Default inventory file created.")

# Edge Case 1: CSV Header Validation
def load_inventory() -> List[Dict]:
    if not os.path.exists("inventory.csv"):
        create_default_csv()

    with file_lock:
        with open("inventory.csv", "r") as f:
            reader = csv.DictReader(f)
            if not reader.fieldnames or set(reader.fieldnames) != {"ProductID", "Name", "Category", "Price", "Quantity", "RestockLevel", "ExpirationDate", "LeadTime"}:
                raise ValueError("Invalid CSV structure. Please check the file.")
            return list(reader)

def save_inventory(inventory: List[Dict]):
    with file_lock:
        with open("inventory.csv", "w", newline='') as f:
            writer = csv.DictWriter(f, fieldnames=["ProductID", "Name", "Category", "Price", "Quantity", "RestockLevel", "ExpirationDate", "LeadTime"])
            writer.writeheader()
            writer.writerows(inventory)

# Edge Case 2: Data Type Checking
# Edge Case 8: Negative Quantities
def validate_product(product: Dict) -> bool:
    if not all(key in product for key in ["ProductID", "Name", "Category", "Price", "Quantity", "RestockLevel", "ExpirationDate", "LeadTime"]):
        return False
    if not re.match(r'^P\d{3}$', product["ProductID"]):
        return False
    try:
        float(product["Price"])
        quantity = int(product["Quantity"])
        int(product["RestockLevel"])
        int(product["LeadTime"])
        if quantity < 0:
            return False
    except ValueError:
        return False
    return True

# Edge Case 3: Duplicate Product IDs
def add_product(inventory: List[Dict]):
    product = {}
    product["ProductID"] = input("Enter Product ID (format: P###): ")
    product["Name"] = input("Name: ")
    product["Category"] = input("Category: ")
    product["Price"] = input("Price: ")
    product["Quantity"] = input("Quantity: ")
    product["RestockLevel"] = input("Restock Level: ")
    product["ExpirationDate"] = input("Expiration Date (YYYY-MM-DD, leave blank if N/A): ")
    product["LeadTime"] = input("Supplier Lead Time (in days): ")

    if not validate_product(product):
        print("Invalid product data. Please check your inputs.")
        return

    if any(p["ProductID"] == product["ProductID"] for p in inventory):
        print("Product ID already exists.")
        return

    inventory.append(product)
    save_inventory(inventory)
    print("Product added successfully.")

def delete_product(inventory: List[Dict]):
    target = input("Enter Product ID to delete: ").strip().upper()
    inventory = [p for p in inventory if p["ProductID"] != target]
    save_inventory(inventory)
    print("Product deleted if it existed.")

# Edge Case 4: Case-Sensitive Search
def search_product(inventory: List[Dict]):
    term = input("Search term: ").lower()
    results = [p for p in inventory if term in p["Name"].lower() or term in p["Category"].lower()]
    if results:
        for product in results:
            print(product)
    else:
        print("No matching products found.")

# Edge Case 5: Restock Level Validation
def auto_restock(inventory: List[Dict]):
    for product in inventory:
        if int(product["Quantity"]) < int(product["RestockLevel"]):
            print(f"Restocking {product['Name']}")
            product["Quantity"] = str(int(product["RestockLevel"]) * 2)
    save_inventory(inventory)
    print("Auto-restock complete.")

# Edge Case 6: Empty File Handling
# Edge Case 7: Category Validation
def generate_report(inventory: List[Dict]):
    if not inventory:
        print("Inventory is empty.")
        return

    categories = {}
    total_value = 0

    for product in inventory:
        total_value += float(product["Price"]) * int(product["Quantity"])
        categories[product["Category"]] = categories.get(product["Category"], 0) + 1

    print(f"Total Inventory Value: ${total_value:.2f}")
    print("Category Breakdown:")
    for cat, count in categories.items():
        print(f"{cat}: {count} items")

# New function to handle seasonal demand fluctuations
def adjust_for_seasonal_demand(inventory: List[Dict]):
    # Simulating seasonal demand (e.g., holiday season)
    is_holiday_season = random.choice([True, False])
    if is_holiday_season:
        print("Adjusting for holiday season demand...")
        for product in inventory:
            product["Quantity"] = str(int(int(product["Quantity"]) * 1.5))  # Increase stock by 50%
        save_inventory(inventory)

# New function to manage product expiration
def manage_expiring_products(inventory: List[Dict]):
    today = datetime.now().date()
    for product in inventory:
        if product["ExpirationDate"]:
            exp_date = datetime.strptime(product["ExpirationDate"], "%Y-%m-%d").date()
            if exp_date - today <= timedelta(days=30):
                print(f"Warning: {product['Name']} is expiring soon. Consider promotions.")

# New function to sync multi-channel inventory
def sync_multi_channel_inventory(inventory: List[Dict]):
    global online_inventory, store_inventory
    for product in inventory:
        online_qty = random.randint(0, int(product["Quantity"]))
        store_qty = int(product["Quantity"]) - online_qty
        online_inventory[product["ProductID"]] = online_qty
        store_inventory[product["ProductID"]] = store_qty
    print("Multi-channel inventory synced.")

# New function to handle returns and damaged goods
def process_returns(inventory: List[Dict]):
    product_id = input("Enter Product ID for return: ")
    condition = input("Enter condition (good/damaged): ").lower()
    for product in inventory:
        if product["ProductID"] == product_id:
            if condition == "good":
                product["Quantity"] = str(int(product["Quantity"]) + 1)
                print("Product returned to inventory.")
            else:
                print("Product marked as damaged. Not returned to inventory.")
            break
    else:
        print("Product not found in inventory.")
    save_inventory(inventory)

# New function to handle bulk orders
def handle_bulk_order(inventory: List[Dict]):
    product_id = input("Enter Product ID for bulk order: ")
    quantity = int(input("Enter order quantity: "))
    for product in inventory:
        if product["ProductID"] == product_id:
            if int(product["Quantity"]) >= quantity:
                product["Quantity"] = str(int(product["Quantity"]) - quantity)
                print("Bulk order processed successfully.")
            else:
                print("Insufficient stock. Order partially fulfilled.")
                product["Quantity"] = "0"
            break
    else:
        print("Product not found in inventory.")
    save_inventory(inventory)

# New function to manage product variants
def manage_product_variants(inventory: List[Dict]):
    base_product_id = input("Enter base Product ID: ")
    variants = []
    while True:
        variant = input("Enter variant (e.g., color, size) or 'done' to finish: ")
        if variant.lower() == 'done':
            break
        variants.append(variant)

    for variant in variants:
        new_product = next((p.copy() for p in inventory if p["ProductID"] == base_product_id), None)
        if new_product:
            new_product["ProductID"] = f"{base_product_id}-{variant}"
            new_product["Name"] = f"{new_product['Name']} ({variant})"
            inventory.append(new_product)

    save_inventory(inventory)
    print("Product variants added to inventory.")

# New function to detect inventory shrinkage
def detect_shrinkage(inventory: List[Dict]):
    for product in inventory:
        expected_quantity = int(product["Quantity"])
        actual_quantity = random.randint(0, expected_quantity)  # Simulating physical count
        if actual_quantity < expected_quantity:
            print(f"Shrinkage detected for {product['Name']}. Expected: {expected_quantity}, Actual: {actual_quantity}")
            product["Quantity"] = str(actual_quantity)
    save_inventory(inventory)

# New function to track supplier lead times
def track_supplier_performance(inventory: List[Dict]):
    for product in inventory:
        actual_lead_time = random.randint(int(product["LeadTime"]) - 5, int(product["LeadTime"]) + 5)
        if actual_lead_time > int(product["LeadTime"]):
            print(f"Supplier delay for {product['Name']}. Expected: {product['LeadTime']} days, Actual: {actual_lead_time} days")
        product["LeadTime"] = str(actual_lead_time)
    save_inventory(inventory)

# New function to manage product bundles
def manage_product_bundles(inventory: List[Dict]):
    bundle_name = input("Enter bundle name: ")
    bundle_products = []
    while True:
        product_id = input("Enter Product ID to add to bundle (or 'done' to finish): ")
        if product_id.lower() == 'done':
            break
        quantity = int(input("Enter quantity for this product in the bundle: "))
        bundle_products.append((product_id, quantity))

    bundle_price = sum(float(next(p["Price"] for p in inventory if p["ProductID"] == pid)) * qty for pid, qty in bundle_products)
    bundle_quantity = min(int(next(p["Quantity"] for p in inventory if p["ProductID"] == pid)) // qty for pid, qty in bundle_products)

    new_bundle = {
        "ProductID": f"B{len([p for p in inventory if p['ProductID'].startswith('B')]) + 1:03d}",
        "Name": bundle_name,
        "Category": "Bundle",
        "Price": str(bundle_price),
        "Quantity": str(bundle_quantity),
        "RestockLevel": "5",
        "ExpirationDate": "",
        "LeadTime": "0"
    }

    inventory.append(new_bundle)
    save_inventory(inventory)
    print(f"Bundle '{bundle_name}' added to inventory.")

# New function to handle cross-location transfers
def handle_cross_location_transfer(inventory: List[Dict]):
    product_id = input("Enter Product ID to transfer: ")
    quantity = int(input("Enter quantity to transfer: "))
    from_location = input("Enter source location (online/store): ").lower()
    to_location = "store" if from_location == "online" else "online"

    global online_inventory, store_inventory
    if from_location == "online":
        if online_inventory.get(product_id, 0) >= quantity:
            online_inventory[product_id] -= quantity
            store_inventory[product_id] = store_inventory.get(product_id, 0) + quantity
            print(f"Transferred {quantity} units from online to store inventory.")
        else:
            print("Insufficient stock in online inventory.")
    else:
        if store_inventory.get(product_id, 0) >= quantity:
            store_inventory[product_id] -= quantity
            online_inventory[product_id] = online_inventory.get(product_id, 0) + quantity
            print(f"Transferred {quantity} units from store to online inventory.")
        else:
            print("Insufficient stock in store inventory.")

def main():
    while True:
        try:
            inventory = load_inventory()
        except Exception as e:
            print(f"Error loading inventory: {e}")
            return

        print("\n1. Add Product\n2. Delete Product\n3. Search\n4. Auto-Restock\n5. Report")
        print("6. Adjust for Seasonal Demand\n7. Manage Expiring Products\n8. Sync Multi-Channel Inventory")
        print("9. Process Returns\n10. Handle Bulk Order\n11. Manage Product Variants")
        print("12. Detect Shrinkage\n13. Track Supplier Performance\n14. Manage Product Bundles")
        print("15. Handle Cross-Location Transfer\n16. Exit")
        choice = input("Select: ")

        try:
            if choice == "1":
                add_product(inventory)
            elif choice == "2":
                delete_product(inventory)
            elif choice == "3":
                search_product(inventory)
            elif choice == "4":
                auto_restock(inventory)
            elif choice == "5":
                generate_report(inventory)
            elif choice == "6":
                adjust_for_seasonal_demand(inventory)
            elif choice == "7":
                manage_expiring_products(inventory)
            elif choice == "8":
                sync_multi_channel_inventory(inventory)
            elif choice == "9":
                process_returns(inventory)
            elif choice == "10":
                handle_bulk_order(inventory)
            elif choice == "11":
                manage_product_variants(inventory)
            elif choice == "12":
                detect_shrinkage(inventory)
            elif choice == "13":
                track_supplier_performance(inventory)
            elif choice == "14":
                manage_product_bundles(inventory)
            elif choice == "15":
                handle_cross_location_transfer(inventory)
            elif choice == "16":
                break
            else:
                print("Invalid choice. Please try again.")
        except Exception as e:
            print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```

```python
import random

flights = []

def add_flight():
    flight_id = input("Enter flight number: ")
    destination = input("Destination: ")
    seats = input("Total seats: ")
    time = input("Departure time: ")

    flights.append({
        'id': flight_id,
        'dest': destination,
        'seats': int(seats),
        'booked': 0,
        'time': time,
        'passengers': []
    })

def find_flights():
    dest = input("Destination: ")
    found = [f for f in flights if dest in f['dest']]
    for f in found:
        print(f"Flight {f['id']} to {f['dest']} at {f['time']}")

def book_seat():
    flight_id = input("Flight number: ")
    for f in flights:
        if f['id'] == flight_id:
            if f['booked'] < f['seats']:
                name = input("Name: ")
                f['passengers'].append(name)
                f['booked'] += 1
                print("Booked!")
            else:
                print("Full")
            return
    print("Flight not found")

def generate_report():
    for f in flights:
        print(f"Flight {f['id']}: {f['booked']}/{f['seats']} booked")

# Initial flights
for _ in range(50):
    flights.append({
        'id': f"FL{random.randint(100,999)}",
        'dest': random.choice(["NYC", "LON", "PAR", "TOK"]),
        'seats': random.randint(50, 200),
        'booked': 0,
        'time': f"{random.randint(0,23):02}:00",
        'passengers': []
    })

while True:
    print("\n1. Add flight\n2. Search\n3. Book\n4. Report\n5. Exit")
    choice = input("Choose: ")

    if choice == "1":
        add_flight()
    elif choice == "2":
        find_flights()
    elif choice == "3":
        book_seat()
    elif choice == "4":
        generate_report()
    elif choice == "5":
        break
```

```python
!pip install bcrypt
import uuid
import bcrypt
from datetime import datetime, timedelta
import re
import threading
import random

class FlightBookingSystem:
    def __init__(self):
        self.flights = []
        self.bookings = {}
        self.users = {}
        self.lock = threading.Lock()
        self.auth_attempts = {}

    # Security Measure 1: Secure PNR Generation
    def generate_secure_pnr(self):
        return str(uuid.uuid4())[:8].upper()

    # Security Measure 2: Input Validation
    def validate_flight_data(self, flight):
        try:
            assert re.match(r'^[A-Z]{2}\d{3,4}$', flight['id'])
            assert re.match(r'^[A-Z]{3}$', flight['origin'])
            assert re.match(r'^[A-Z]{3}$', flight['destination'])
            assert 0 < int(flight['seats']) <= 1000
            datetime.strptime(flight['time'], '%Y-%m-%d %H:%M')
            return True
        except (AssertionError, ValueError):
            return False

    # Security Measure 3: Brute-force Protection
    def check_auth_attempt(self, username):
        with self.lock:
            if username in self.auth_attempts:
                if self.auth_attempts[username]['count'] >= 5:
                    if datetime.now() - self.auth_attempts[username]['last_attempt'] < timedelta(minutes=15):
                        return False
                    else:
                        self.auth_attempts[username]['count'] = 0
            return True

    # Security Measure 4: Secure Data Storage
    def add_flight(self, flight):
        if not self.validate_flight_data(flight):
            print("Invalid flight data")
            return False
        with self.lock:
            if any(f['id'] == flight['id'] for f in self.flights):
                print("Flight ID already exists")
                return False
            self.flights.append(flight)
            return True

    # Security Measure 5: Prevent Overbooking
    def book_seat(self, flight_id, username):
        with self.lock:
            flight = next((f for f in self.flights if f['id'] == flight_id), None)
            if not flight or len(self.bookings.get(flight_id, [])) >= int(flight['seats']):
                print("Flight not found or fully booked")
                return None
            pnr = self.generate_secure_pnr()
            self.bookings.setdefault(flight_id, []).append({'pnr': pnr, 'user': username})
            return pnr

    # Security Measure 6: User Authentication
    def register_user(self, username, password):
        if username in self.users:
            print("Username already exists")
            return False
        hashed = bcrypt.hashpw(password.encode(), bcrypt.gensalt())
        self.users[username] = hashed
        return True

    def authenticate_user(self, username, password):
        if not self.check_auth_attempt(username):
            print("Too many failed attempts. Try again later.")
            return False
        if username not in self.users:
            print("User not found")
            return False
        if bcrypt.checkpw(password.encode(), self.users[username]):
            self.auth_attempts.pop(username, None)
            return True
        with self.lock:
            self.auth_attempts.setdefault(username, {'count': 0, 'last_attempt': datetime.now()})
            self.auth_attempts[username]['count'] += 1
            self.auth_attempts[username]['last_attempt'] = datetime.now()
        print("Invalid password")
        return False

    # Security Measure 7: Logging
    def log_activity(self, activity):
        with open('activity_log.txt', 'a') as f:
            f.write(f"{datetime.now()}: {activity}\n")

    # Main loop
    def run(self):
        while True:
            print("\n1. Register\n2. Login\n3. Exit")
            choice = input("Choose an option: ")
            if choice == '1':
                username = input("Enter username: ")
                password = input("Enter password: ")
                if self.register_user(username, password):
                    print("User registered successfully")
                    self.log_activity(f"User registered: {username}")
            elif choice == '2':
                username = input("Enter username: ")
                password = input("Enter password: ")
                if self.authenticate_user(username, password):
                    print("Login successful")
                    self.log_activity(f"User logged in: {username}")
                    self.user_menu(username)
            elif choice == '3':
                print("Exiting...")
                break
            else:
                print("Invalid choice")

    def user_menu(self, username):
        while True:
            print("\n1. Add Flight\n2. Search Flights\n3. Book Flight\n4. My Bookings\n5. Logout")
            choice = input("Choose an option: ")
            if choice == '1':
                flight = {
                    'id': input("Flight ID (e.g., AA123): "),
                    'origin': input("Origin (3-letter code): "),
                    'destination': input("Destination (3-letter code): "),
                    'seats': input("Total seats: "),
                    'time': input("Departure time (YYYY-MM-DD HH:MM): ")
                }
                if self.add_flight(flight):
                    print("Flight added successfully")
                    self.log_activity(f"Flight added: {flight['id']} by {username}")
            elif choice == '2':
                origin = input("Origin: ").upper()
                destination = input("Destination: ").upper()
                found = [f for f in self.flights if f['origin'] == origin and f['destination'] == destination]
                if found:
                    for f in found:
                        print(f"Flight {f['id']} from {f['origin']} to {f['destination']} at {f['time']}")
                else:
                    print("No flights found")
            elif choice == '3':
                flight_id = input("Enter Flight ID: ")
                pnr = self.book_seat(flight_id, username)
                if pnr:
                    print(f"Booking successful. Your PNR is {pnr}")
                    self.log_activity(f"Booking made: {flight_id} by {username}")
            elif choice == '4':
                user_bookings = []
                for flight_id, bookings in self.bookings.items():
                    for booking in bookings:
                        if booking['user'] == username:
                            flight = next(f for f in self.flights if f['id'] == flight_id)
                            user_bookings.append(f"PNR: {booking['pnr']}, Flight: {flight_id}, From: {flight['origin']} To: {flight['destination']} at {flight['time']}")
                if user_bookings:
                    print("Your bookings:")
                    for booking in user_bookings:
                        print(booking)
                else:
                    print("No bookings found")
            elif choice == '5':
                print("Logging out...")
                self.log_activity(f"User logged out: {username}")
                break
            else:
                print("Invalid choice")

# Initialize the system with some flights
system = FlightBookingSystem()
for _ in range(50):
    flight = {
        'id': f"AA{random.randint(100,999)}",
        'origin': random.choice(["JFK", "LAX", "ORD", "DFW", "ATL"]),
        'destination': random.choice(["LHR", "CDG", "FRA", "AMS", "MAD"]),
        'seats': str(random.randint(150, 300)),
        'time': (datetime.now() + timedelta(days=random.randint(1,30))).strftime('%Y-%m-%d %H:%M')
    }
    system.add_flight(flight)

# Run the system
system.run()
```
